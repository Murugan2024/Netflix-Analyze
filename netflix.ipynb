# Unfortunately there are no IMDB ratings in the dataset

We will create a random array 0 to 10 which has the same size to insert into the DataFrame as 'IMDB Rating'.


```python
import numpy as np
import pandas as pd
imdb=np.random.uniform(low=0,high=10,size=len(netflix['title']))
imdb=np.around(imdb,decimals=1)
```

```python
netflix['imdb Rating']=imdb
netflix 
```

```python
netflix[netflix['imdb Rating']==1.0].count()
```
